```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("../dataset/metadata/test/test.csv", sep="\t")
```

```python
df["duration"] = df["offset"] - df ["onset"]
```

```python
f, arr = plt.subplots(2,5, figsize=(20,10))
for i, label in enumerate(np.unique(df["event_label"])):
    df[df["event_label"] == label]["duration"].hist(bins=20, ax=arr[int(i/5), i%5])
    arr[int(i/5), i%5].set_title(label)
    arr[int(i/5), i%5].set_xlabel("duration")
    if i%5 == 0:
        arr[int(i/5), i%5].set_ylabel("counts")
plt.show()
```

### Mean duration of events

```python
means = []
labels = np.unique(df["event_label"])
print("Mean duration of events")
for label in labels:
    mean_duration = df[df["event_label"] == label]["duration"].mean()
    means.append(mean_duration)
    print("{:30} : {:3f}".format(label, mean_duration))
```

### Median duration of events

```python
medians = []
labels = np.unique(df["event_label"])
print("Median duration of events")
for label in labels:
    mean_duration = df[df["event_label"] == label]["duration"].median()
    medians.append(mean_duration)
    print("{:30} : {:3f}".format(label, mean_duration))
```

### Durations, mean over events

```python
print("mean duration on mean events (macro)")
print(np.mean(means))
```

```python
print("median duration on median events (macro)")
print(np.median(medians))
```

### Durations, overall data

```python
print("Median duration over all events (micro)")
print(df["duration"].median())
```

```python
print("mean duration over all events (micro)")
df["duration"].mean()
```
